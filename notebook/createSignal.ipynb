## Just a notebook to create some signals and download them

```python
import numpy as np
from scipy import signal
import IPython.display as ipd
```

```python
sr = 22050
def createSound(hz, amplitude = 1, T=2, sr=22050):
    t = np.linspace(0, T, int(T*sr), endpoint=False)
    x= np.abs(signal.sawtooth(1 * np.pi * 100 * t)) 
    #x = signal.square(amplitude*np.sin(2*np.pi*hz*t))
    return x

x = createSound(100)
ipd.Audio(x, rate=sr)
```
